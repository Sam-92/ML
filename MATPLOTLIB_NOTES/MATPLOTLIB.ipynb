## LINE GRAPH

```python
import numpy as np
```

```python
x=np.array([1,3,6,8])
y=np.array([2,7,8,9]);
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.show()
```

### SETTING TITLES , AND WIDTH AND STYLING

```python
from matplotlib import style

style.use("ggplot")

plt.plot(x,y,color="green",linewidth=5,label="line_one")
plt.title("info")
plt.ylabel("Y-AXIS")
plt.xlabel("X-AXIS")
plt.legend()
plt.grid(True,axis="both",color="blue")
plt.show()
```

## BAR GRAPH

```python
plt.bar(x,y,color="blue",label="Example one")
plt.legend()
plt.title("BAR-GRAPH")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#### bar graph uses categorical variables while histogram uses quantitative variables

## HISTOGRAM

```python
x=[1,2,3,4,6,7,9,0]
bins=[2,4,6,8];
plt.hist(x,bins,histtype="bar",rwidth=0.3)
plt.show()
```

## SCATTER-PLOT

```python
x=[2,54,7,9,7]
y=[1,2,3,4,5]
plt.scatter(x,y,color="blue",marker="o",s=25)

plt.show()
```

## PLOTTING OUR OWN GRAPH AND USING SUBPLOT

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(2,1,1)
plt.plot(t1,f(t1),"o",color="green")

plt.subplot(2,1,2)
plt.plot(t2,np.cos(2*np.pi*t2),"b2")
```
