# ITERATING THROUGH DATAFRAMES AND SERIES

```python
import pandas as pd
df=pd.read_csv("pandas-videos-master/data/ufo.csv")
df.head()
```

## ITERATING THROUGH SERIES

```python
for i in df["City"]:
    print(i)
```

## ITERATING THROUGH DATAFRAMES

### USING ITERROWS::

```python
for index,row in df.iterrows():
    print(index,row["State"],row["City"])
```
