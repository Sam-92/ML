## WORKING WITH THE FAMOUS TITANIC DATA

### READING THE DATA

```python
# Reading the data
import pandas as pd
titanic=pd.read_csv("train.csv")
titanic.head()
```

```python
titanic.dtypes
```

```python
titanic.Ticket.value_counts()
```

```python
titanic.info()
```

### FINDING MISSING VALUES

```python
titanic.isnull().sum()
```

```python
titanic.drop(labels="Cabin",axis=1,inplace=True)
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.isnull().sum()
```

### REPLACING MISSING VALUES WITH MEAN OR MOST FREQUENT OCCURING VALUE

```python
import numpy as np
titanic.Age.fillna(value=titanic.Age.mean(),inplace=True)
titanic.Age.isnull().sum()
```

```python
titanic.Embarked.fillna(value=titanic.Embarked.describe().top,inplace=True)
titanic.Embarked.isnull().sum()
```

```python
titanic.describe()
```

```python
target=titanic.Survived
target
```

```python
titanic.head()
```

```python
titanic.drop(['Name','Survived','Ticket'],axis=1,inplace=True)
```

### NORMALIZING FARES

```python
titanic["Fare"] = titanic["Fare"]/titanic.Fare.max()
titanic.head()
```

### ONE HOT ENCODING EMBARKED AND SEX COLUMNS

```python
dummy_Embarked = pd.get_dummies(titanic.Embarked,prefix="Embarked",drop_first=True)
dummy_Sex= pd.get_dummies(titanic.Sex,prefix="Sex",drop_first=True)
dummy_Embarked.head()
```

```python
titanic=pd.concat([titanic,dummy_Embarked],axis=1)
titanic.head()
```

```python
titanic=pd.concat([titanic,dummy_Sex],axis=1)
titanic.head()
```

```python
titanic.drop(['Embarked','Sex'],axis=1,inplace=True)
titanic.head()
```

### READING IN THE TEST DATA AND MAKING IT SIMILAR TO TRAINING SET

```python
test=pd.read_csv("test.csv")
test.head()
```

```python
test.isnull().sum()
```

```python
test.drop(["Name","Cabin"],axis=1,inplace=True)
```

```python
test.head()
```

```python
test.Age.fillna(value=test.Age.mean(),inplace=True)
test.Fare.fillna(value=test.Fare.mean(),inplace=True)
test.isnull().sum()
```

```python
test.describe()
test["Fare"] = test.Fare/test.Fare.max()
test.describe()
```

```python
dummy_test_sex = pd.get_dummies(test.Sex,prefix="Sex",drop_first=True)
dummy_test_sex.head()
```

```python
dummy_test_Embarked = pd.get_dummies(test.Embarked,prefix="Embarked",drop_first=True)
dummy_test_Embarked.head()
```

```python
test=pd.concat([test,dummy_test_Embarked,dummy_test_sex],axis=1)
```

```python
test.drop(["Embarked","Sex"],axis=1,inplace=True)
```

```python
test.drop("Ticket",axis=1,inplace=True)
```

## APPLYING CLASSIFICATION MODELS

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
```

```python
clf11 = LogisticRegression()
clf21 = SVC()
clf31 = tree.DecisionTreeClassifier()
clf41 = RandomForestClassifier(n_estimators=60)

clf12 = LogisticRegression()
clf22 = SVC()
clf32 = tree.DecisionTreeClassifier()
clf42 = RandomForestClassifier(n_estimators=60)
```

## USING HOLDOUT METHOD

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(titanic,target,test_size=0.4,random_state=42)
```

### IMPLEMENTING LOGISTIC REGRESSION

```python
clf11.fit(X_train,y_train)
clf11.score(X_test,y_test)
```

### IMPLEMENTING SVM

```python
clf21.fit(X_train,y_train)
clf21.score(X_test,y_test)
```

### IMPLEMENTING DECISION TREE

```python
clf31.fit(X_train,y_train)
clf31.score(X_test,y_test)
```

### IMPLEMENTING RANDOM FOREST CLASSIFICATION

```python
clf41.fit(X_train,y_train)
clf41.score(X_test,y_test)
```

#### MAKING DATAFRAME FOR PASSENGERID AND SURVIVED PASSENGERS

```python
y_pred11=clf11.predict(test)
y_pred21=clf21.predict(test)
y_pred31=clf31.predict(test)
y_pred41=clf41.predict(test)
```

```python
df11=pd.DataFrame({"PassengerId":test.PassengerId , "Survived":y_pred11}).set_index("PassengerId")
df21=pd.DataFrame({"PassengerId":test.PassengerId , "Survived":y_pred21}).set_index("PassengerId")
df31=pd.DataFrame({"PassengerId":test.PassengerId , "Survived":y_pred31}).set_index("PassengerId")
df41=pd.DataFrame({"PassengerId":test.PassengerId , "Survived":y_pred41}).set_index("PassengerId")
```

#### WRITING TO CSV FILE

```python
df11.to_csv("holdout_logisticRegression.csv")
df21.to_csv("holdout_SVM.csv")
df31.to_csv("holdout_DecisionTree.csv")
df41.to_csv("holdout_RandomForest.csv")
```

## USING KFOLD CROSS VALIDATION

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
```

### IMPLEMENTING LOGISTIC REGRESSION

```python
scores=[]
for train_index,test_index in kf.split(titanic,target):
    X_train=titanic.loc[train_index,:]
    X_test=titanic.loc[test_index,:]
    y_train=target[train_index]
    y_test=target[test_index]
    clf12.fit(X_train,y_train)
    scores.append(clf12.score(X_test,y_test))
print(np.mean(scores))
```

### IMPLEMENTING SVM

```python
scores=[]
for train_index,test_index in kf.split(titanic,target):
    X_train=titanic.loc[train_index,:]
    X_test=titanic.loc[test_index,:]
    y_train=target[train_index]
    y_test=target[test_index]
    clf22.fit(X_train,y_train)
    scores.append(clf22.score(X_test,y_test))
print(np.mean(scores))
```

### IMPLEMENTING DECISION TREE

```python
scores=[]
for train_index,test_index in kf.split(titanic,target):
    X_train=titanic.loc[train_index,:]
    X_test=titanic.loc[test_index,:]
    y_train=target[train_index]
    y_test=target[test_index]
    clf32.fit(X_train,y_train)
    scores.append(clf32.score(X_test,y_test))
print(np.mean(scores))
```

### IMPLEMENTING RANDOM FOREST CLASSIFICATION

```python
scores=[]
for train_index,test_index in kf.split(titanic,target):
    X_train=titanic.loc[train_index,:]
    X_test=titanic.loc[test_index,:]
    y_train=target[train_index]
    y_test=target[test_index]
    clf42.fit(X_train,y_train)
    scores.append(clf42.score(X_test,y_test))
print(np.mean(scores))
```

#### MAKING DATAFRAME FOR PASSENGERID AND SURVIVED PASSENGERS

```python
y_pred12=clf12.predict(test)
y_pred22=clf22.predict(test)
y_pred32=clf32.predict(test)
y_pred42=clf42.predict(test)
```

```python
df12=pd.DataFrame({"PassengerId":test.PassengerId , "Survived":y_pred12}).set_index("PassengerId")
df22=pd.DataFrame({"PassengerId":test.PassengerId , "Survived":y_pred22}).set_index("PassengerId")
df32=pd.DataFrame({"PassengerId":test.PassengerId , "Survived":y_pred32}).set_index("PassengerId")
df42=pd.DataFrame({"PassengerId":test.PassengerId , "Survived":y_pred42}).set_index("PassengerId")
```

#### WRITING TO CSV FILE

```python
df12.to_csv("Kfold_logisticRegression.csv")
df22.to_csv("Kfold_SVM.csv")
df32.to_csv("Kfold_DecisionTree.csv")
df42.to_csv("Kfold_RandomForest.csv")
```
