# SORTING PANDASERIES OR DATAFRAME

```python
import pandas as pd
```

```python
movies=pd.read_csv("pandas-videos-master/data/imdb_1000.csv")
movies.head()
```

## SORTING PANDA SERIES

### Sorting in ascending order

```python
movies['title'].sort_values()
```

### Sorting in descending order

```python
movies['title'].sort_values(ascending=False)
```

## SORTING DATAFRAME

```python
movies.sort_values('title')
```

```python
df =movies.sort_values('duration',ascending=False)
df.head()
```

### We can sort dataframe with respect to multiple coloumns..It will be sorted first on the basis of first entry and then on the basis of second 

```python
df=movies.sort_values(['title','duration'],ascending=False)
df.head()
```
