# LOC

```python
import pandas as pd
```

```python
ufo=pd.read_csv("pandas-videos-master/data/ufo.csv")
ufo.head()
```

## USING LOC

### : REFERS ALL

```python
ufo.loc[0 , :]
```

```python
ufo.loc[[0,1,2],:]
```

```python
ufo.loc[0:2,:]
```

```python
ufo.loc[0:2,:]
```

```python
ufo.loc[0:2 , ["City","Colors Reported","Shape Reported"]]
```

```python
ufo.loc[0:2 ,"City":"Shape Reported"]
```

## WE CAN ALSO FILTER ROWS

```python
ufo.loc[ufo.State=="NY",:]
```

### or

```python
ufo[ufo.State == "NY"]
```

```python
ufo.loc[ufo.State=="NY","City":"Shape Reported"]
```
